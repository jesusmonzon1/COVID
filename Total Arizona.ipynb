```python
import pandas as pd
import numpy as np

df = pd.read_csv('time_series_covid19_confirmed_US.csv')
```

```python
# yuma = (df['Admin2'] == 'Yuma') & (df['Province_State'] == 'Arizona')
az = (df['Province_State'] == 'Arizona')
df2 = df.loc[az]
```

```python
df2.rename(columns={'Admin2':'Dates'}, inplace=True)
df2.set_index('Dates', inplace=True)
df2 = df2.T
```

```python
df2.drop(['UID', 'iso2', 'iso3', 'code3', 'FIPS',
          'Province_State', 'Country_Region', 'Lat', 'Long_', 'Combined_Key'], inplace=True)
```

```python
df2['Date'] = df2.index
df2['Date'] = pd.to_datetime(df2['Date'], format= '%m/%d/%y')
df2.set_index('Date', inplace=True)
```

```python
# df2
```

```python
df2 = df2[df2.index.month >= 3]
```

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

plt.style.use('fivethirtyeight')

# Create figure and plot space
fig, ax = plt.subplots(figsize=(10, 10))

# Add lines
ax.plot(df2.index.values,
        df2['Apache'], linewidth=2, label='Apache')

ax.plot(df2.index.values,
        df2['Cochise'], linewidth=2, label='Cochise')

ax.plot(df2.index.values,
        df2['Gila'], linewidth=2, label='Gila')

ax.plot(df2.index.values,
        df2['Graham'], linewidth=2, label='Graham')

ax.plot(df2.index.values,
        df2['La Paz'], linewidth=2, label='La Paz')

# ax.plot(df2.index.values,
#         df2['Maricopa'], linewidth=2, label='Maricopa')

ax.plot(df2.index.values,
        df2['Mohave'], linewidth=2, label='Mohave')

ax.plot(df2.index.values,
        df2['Navajo'], linewidth=2, label='Navajo')

ax.plot(df2.index.values,
         df2['Pima'], linewidth=2, label='Pima')

ax.plot(df2.index.values,
         df2['Pinal'], linewidth=2, label='Pinal')

ax.plot(df2.index.values,
         df2['Santa Cruz'], linewidth=2, label='Santa Cruz')

ax.plot(df2.index.values,
         df2['Yavapai'], linewidth=2, label='Yavapai')

ax.plot(df2.index.values,
        df2['Yuma'], linewidth=2, label='Yuma')




# Set title and labels for axes
ax.set(xlabel="DATE",
       ylabel="NUMBER OF INDIVIDUALS",
       title="CASES CONFIRMED IN ARIZONA BY COUNTY")

ax.legend()

plt.tight_layout()
plt.show()
```
